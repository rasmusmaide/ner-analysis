```python
kasitsi = pd.read_csv("corpora/estner.cnll", delimiter="\t", names = ["word", "lemma", "form", "nertag"])
```

```python
kasitsi = kasitsi.dropna()
```

```python
kasitsi.lemma = kasitsi.lemma.apply(lambda x: x.split("+")[0])
```

```python
kasilemmad = []
```

```python
entity_spans = []
entity_type = None
for span, label in zip(kasitsi.lemma, kasitsi.nertag):
    if entity_type is None:
        entity_type = label[2:]
    if label == "O":
        if entity_spans:
            kasilemmad.append({"Lemma": " ".join(entity_spans), "Tag": entity_type})
            entity_spans = []
            continue
    if label[0] == "B" or entity_type != label[2:]:
        if entity_spans:
            kasilemmad.append({"Lemma": " ".join(entity_spans), "Tag": entity_type})
            entity_spans = []
    entity_type = label[2:]
    entity_spans.append(span)
```

```python
kasidf = pd.DataFrame(kasilemmad)
```

```python
kasidf = kasidf[kasidf.Tag != ""]
```

```python
kasidf = kasidf.reset_index(drop=True)
```

```python
duplicated = pd.read_csv("all_duplicates.csv")
```

```python
duplicated = duplicated.sort_values('0',ascending=False).reset_index(drop=True).drop("Unnamed: 0",axis=1)
```

```python
duplicated['0'][320]
```

```python
duplicated
```

```python
yhesed = []
for i in range(320):
    count = duplicated['0'][i]
    tekst = duplicated['lemma'][i]
    kokku = duplicated[duplicated['lemma']==tekst]['0'].sum()
    if count / kokku > 0.9:
        yhesed.append({'text': tekst,'tag': duplicated['tag'][i]})
```

```python
yhesed
```

```python
locs = []
orgs = []
pers = []

for row in yhesed:
    if row['tag'] == 'LOC':
        locs.append(row['text'])
    elif row['tag'] == 'ORG':
        orgs.append(row['text'])
    else:
        pers.append(row['text'])
```

```python
duplicated[duplicated['lemma']=='Eesti'][duplicated['tag']!='LOC']
```

```python
changed_locs = 0
changed_orgs = 0
changed_pers = 0
for loc in locs:
    loccount = duplicated[duplicated['lemma']==loc][duplicated['tag']!='LOC']['0'].sum()
    changed_locs += loccount
for org in orgs:
    orgcount = duplicated[duplicated['lemma']==loc][duplicated['tag']!='LOC']['0'].sum()
    changed_orgs += orgcount
for per in pers:
    percount = duplicated[duplicated['lemma']==loc][duplicated['tag']!='LOC']['0'].sum()
    changed_pers += percount
```

```python
changed_locs+changed_orgs+changed_pers
```
